```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
bankdf = pd.read_csv('BankChurners.csv')
bankdf.head()
```

```python
bankdf.info()
```

```python
import seaborn as sns
plt.figure(figsize=(6,4))
sns.countplot(data=bankdf, x='Attrition_Flag')
plt.title("Distribution of Customer Churn")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=bankdf, x='Customer_Age', hue='Attrition_Flag', kde=True, bins=20)
plt.title("Age Distribution by Churn Status")
plt.show()
```
